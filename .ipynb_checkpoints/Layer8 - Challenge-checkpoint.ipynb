#### Missing values

```python
# Absolute quantity of NaN values for each column

raw_data.isna().sum()
```

```python
# Percentage of NaN values for each column

raw_data.isna().sum()/len(raw_data)*100
```

```python
# Dropping rows with missing data in columns 'name' and 'usd pledged'
# Chosen to remove all NaN data (in this case raw_data['status'] == 'undefined' was also removed)

raw_data_M = raw_data.dropna(subset=['name', 'usd pledged']).reset_index(drop=True)
raw_data.isna().sum()
```

```python
# Checking absence of 'undefined' in status column

raw_data_M['state'].unique()
```

#### Duplicates

```python
# We have duplicate projects names
# I decided to maintain because they are differents projects that have differents states

duplicated_names = raw_data_M.loc[raw_data_M.name.duplicated(keep=False), :]
```

```python
# No duplicate names and launched dates

print(duplicated_names[duplicated_names.duplicated(subset=["name", "launched"], keep=False)].shape)
display(duplicated_names[duplicated_names.duplicated(subset=["name", "launched"], keep=False)])
```

### Bivariate Analysis


```python
corr_matrix = clean_data.corr()
sns.heatmap(corr_matrix, cmap='bwr', annot=True);
```

```python
# Converting the correlation matrix into a DataFrame
corr_df = corr_matrix.stack().reset_index()

# Renaming the columns
corr_df.columns = ['feature_1','feature_2', 'correlation']

# Remove "self correlations"
no_self_correlation = (corr_df['feature_1'] != corr_df['feature_2'])
corr_df = corr_df[no_self_correlation]

# Computing the absolute correlation
corr_df['absolute_correlation'] = np.abs(corr_df['correlation'])

# Showing some correlated pairs of feature
corr_df.sort_values(by="absolute_correlation", ascending=False).head(20)
```

```python
real_state = label_encoder.inverse_transform(clean_data['encoded_state'])
clean_data['real_state'] = real_state
state_pledged = clean_data.groupby(['real_state'], as_index=False).agg({'goal':pd.Series.count})
state_pledged['%'] = (state_pledged['goal'] / state_pledged['goal'].sum()*100)
state_pledged.columns = ['state', 'nr_projects', '%']
state_pledged.sort_values(by=['nr_projects'], ascending=False)
```

```python
country_state = clean_data.groupby(['country', 'encoded_state'], as_index=False).agg({'ID': pd.Series.count})
country_state.sort_values('ID', ascending=False, inplace=True) # Se colocar inplace=True, a ordenação decrescente é gravada na variável cat_sex e interfere na forma que o gráfico é apresentado.
```

```python
graph_country_state = px.bar(x=country_state['country'], y=country_state['ID'], color=label_encoder.inverse_transform(country_state['encoded_state']))

graph_country_state.update_layout(xaxis_title='Countries', yaxis_title='Projects per country')

graph_country_state.show()
```
